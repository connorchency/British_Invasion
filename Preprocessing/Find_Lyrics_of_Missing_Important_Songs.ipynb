```python
q = '''
    select a.artist, b.song, b.text from missing a join songdata b on a.song = b.song group by a.artist, b.song, b.text
    '''
result = ps.sqldf(q, globals())
```

```python
ID = []
for i in range(len(result)):
    ID.append(i+1)
```

```python
result["ID"] = ID
```

```python
q1 = '''select artist, song, lyrics from(
    select artist, song, text as lyrics, max(ID)
    from result
    group by artist, song)
    group by artist, song, lyrics
    '''
out = ps.sqldf(q1, globals())
```
