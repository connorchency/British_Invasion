```python
import billboard, json
from datetime import timedelta, datetime
```

```python
chartDate = []
charts = []
chartDate.append('1964-01-04')
charts.append(billboard.ChartData('hot-100', '1964-01-04'))
for i in range(1,313):
    chartDate.append((datetime.strptime(chartDate[i-1], '%Y-%m-%d') 
                      + timedelta(weeks = 1)).strftime('%Y-%m-%d'))
    #print chartDate[i]
    charts.append(billboard.ChartData('hot-100', str(chartDate[i])))
```

```python
# chartDate[311]

# bbTable = {}
# bb_tmp = {'artist': None, 'rank': None}
# #artist = []
# #rank = []

# artists = []
# songs = []
# stay = []
# rank = []

bb_wk_list = []
for i in range(len(charts)):
    ch = charts[i]
    for j in range(100):
        bb_tmp = {}
        bb_tmp['week'] = bb_tmp.get('week', chartDate[i])
        bb_tmp['artist'] = bb_tmp.get('artist', ch[j].artist)
        bb_tmp['rank'] = bb_tmp.get('rank', ch[j].rank)
        bb_tmp['stay'] = bb_tmp.get('stay', ch[j].weeks)
        bb_tmp['song'] = bb_tmp.get('song', ch[j].title)
        bb_wk_list.append(bb_tmp)
    
```

```python
import pandas as pd
df = pd.DataFrame.from_dict(bb_wk_list)
df.to_csv('')
```
